```python
df = pd.read_csv("/content/delhi.csv")
df.head()
```

```python
import pandas as pd

# Define holidays
holidays = pd.DataFrame({
    'holiday': ['Anamoly'],
    'ds': ['2022-06-01']
})

# Convert 'ds' to datetime format
holidays['ds'] = pd.to_datetime(holidays['ds'])
```

```python
import pandas as pd
from prophet import Prophet

# Assuming 'delhi' is your DataFrame
# Reset the index to bring 'Month' as a column
#delhi = delhi.reset_index()

# Prepare data for Prophet
df.rename(columns={'Month': 'ds', 'energy_requirement': 'y'}, inplace=True)

# Initialize Prophet model
model = Prophet(holidays=holidays)

# Add 'tavg' as an additional regressor
model.add_regressor('tmax')

# Fit the model
model.fit(df[['ds', 'y', 'tmax']])

# Make future predictions (you can create a dataframe with future dates)
future = model.make_future_dataframe(periods=12, freq='MS')  # MS for month start

# Ensure you have 'tavg' values for all future dates in the 'future' dataframe
# If you don't have future 'tavg' values, consider using a different method to fill them,
# such as using historical averages or another forecasting model.
# Here's an example of filling with historical averages:
# Convert 'ds' column in both DataFrames to datetime
future['ds'] = pd.to_datetime(future['ds'])
df['ds'] = pd.to_datetime(df['ds'])

future = future.merge(df[['ds', 'tmax']], on='ds', how='left')
future['tmax'].fillna(df['tmax'].mean(), inplace=True)

# Predict
forecast = model.predict(future)

# Show the forecast
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
```

```python
import matplotlib.pyplot as plt

# Plotting the forecast with matplotlib
plt.figure(figsize=(10, 6))

# Plot the forecasted values
plt.plot(forecast['ds'], forecast['yhat'], label='Forecasted Energy Requirement')

# Plot the lower and upper bounds for the forecast
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='gray', alpha=0.3, label='Confidence Interval')

# Add titles and labels
plt.title('Energy Requirement Forecast')
plt.xlabel('Date')
plt.ylabel('Energy Requirement')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the actual vs predicted with confidence interval using matplotlib
plt.figure(figsize=(10, 6))

# Plot actual values
plt.plot(df['ds'], df['y'], label='Actual Energy Requirement', color='blue')

# Plot predicted values
plt.plot(forecast['ds'], forecast['yhat'], label='Predicted Energy Requirement', color='orange')

# Plot the lower and upper bounds for the forecast
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='gray', alpha=0.3, label='Confidence Interval')

# Add titles and labels
plt.title('Energy Requirement: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Energy Requirement')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import plotly.express as px
from datetime import datetime

# Load and prepare the data
df = pd.read_csv('/content/delhi_peak.csv')  # Replace with your CSV file path
df['Month'] = pd.to_datetime(df['Month'])
df.rename(columns={'Month': 'ds', 'peak_demand': 'y'}, inplace=True)

# Prepare DataFrame for Prophet
# Assuming df contains columns 'ds' (dates), 'y' (peak_demand), and 'tmax' (temperature max)
model = Prophet()
model.add_regressor('tmax')  # Add tmax as an additional regressor

# Fit the model
model.fit(df[['ds', 'y', 'tmax']])

# Create a future DataFrame
future = model.make_future_dataframe(periods=12, freq='MS')  # Forecast for 12 months ahead

# Assuming you have future tmax data for the forecast period
# For demonstration, we'll use historical tmax values; replace this with actual future data
future = future.merge(df[['ds', 'tmax']], on='ds', how='left')
future['tmax'] = future['tmax'].fillna(df['tmax'].mean())  # Fill missing values with historical mean

# Predict future values
forecast = model.predict(future)

# Plot forecast results
fig = px.line(forecast, x='ds', y='yhat', title='Forecasted Peak Demand')
fig.update_traces(mode='lines+markers')
fig.show()

# Print forecast results
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```
